```python
import numpy as np
import matplotlib.pyplot as plt


plt.rcParams['figure.dpi'] = 150
```

## $\begin{cases}
       \frac{dx}{dt} = a \cdot x - b \cdot xy\\
       \frac{dy}{dt} = c \cdot xy - d \cdot y,
    \end{cases}$
##    a, b, c, d > 0

```python
PreyAmount = 5.1
PredatorAmount = 4.9
a = 10
b = 2
c = 2
d = 10
dt = 1 / 1000
state_point = (a / b, d / c)
T = 10
t = np.arange(0, T, dt)
u = np.empty((t.size, 2))
u[0] = np.array([PreyAmount, PredatorAmount])
f = lambda x, y: np.array([x * a - b * x * y, y * c * x - d * y])
```

```python
for i in range(t.size - 1):
	alpha = 3 / 4
	u[i + 1] = u[i] + dt * ((1 - alpha) * f(*u[i]) + alpha * f(*u[i] + dt / (2 * alpha) * f(*u[i])))
x = u[:, 0]
y = u[:, 1]
```

## Solution

```python
plt.plot(t, x, ':', c='r', label='Preys')
plt.plot(t, y, c='k', label='Predators')
plt.xlabel('Time')
plt.ylabel('Number of creatures')
plt.legend(loc='upper right')
plt.show()
```

## Phase trajectory

```python
plt.scatter(PreyAmount, PredatorAmount, c='r', zorder=1, label='Initial amount of preys and predators')
plt.plot(x, y, c='y', zorder=0)
plt.xlabel('Preys')
plt.ylabel('Predators')
plt.legend(loc='upper right')
plt.show()
```
